**"_Was sind die molekularen Unterschiede zwischen Krebszellen und normalem menschlichem Gewebe, und wie können wir diese Unterschiede nutzen, um neue therapeutische Ziele und deren Medikamente für die Krebsbehandlung zu identifizieren?_"**

### **Warum dies für die Medizin wichtig ist:**

- **Therapeutische Ziele:** Neue Medikamente werden dringend für bessere Patientenergebnisse benötigt

### **Was wir entdecken werden:**
1. **Molekulare Krebssignaturen:** Gene, die konsistent in Krebs vs. normalem Gewebe verändert sind
2. **Therapeutische Schwachstellen:** Signalwege, die mit Medikamenten angesteuert werden könnten
3. **Drug Repositioning-Möglichkeiten:** Bestehende Medikamente, die Krebs behandeln könnten
4. **Biomarker-Identifikation:** Gene, die das Ansprechen auf die Behandlung vorhersagen könnten

## Datensatz: Eine echte Studie Krebs vs. Normalgewebe

**Klinischer Kontext:** Universal Human Reference (UHR) vs Human Brain Reference (HBR)  
**Medizinische Relevanz:** Krebszelllinien vs. normales menschliches Hirngewebe<br>
**Stichprobengröße:** 6 Proben (3 Krebs-Replikate vs. 3 Normal-Hirngewebe-Replikate)  
**Datentyp:** Paired-End-RNA-Sequenzierung (Chromosom 22 Subset)  
**Referenz:** Griffith M, Walker JR, Spies NC, Ainscough BJ, Griffith OL (2015) Informatics for RNA Sequencing: A Web Resource for Analysis on the Cloud. *PLoS Comput Biol* 11(8): e1004393. https://doi.org/10.1371/journal.pcbi.1004393



### **Medizinische Proben**

#### **UHR (Universal Human Reference) = KREBSPROBEN**
- **Zusammensetzung:** Gesamt-RNA aus 10 verschiedenen humanen Krebszelllinien
- **Enthaltene Krebsarten:** Brust-, Leber-, Gebärmutterhals-, Hoden-, Hirn-, Hautkrebsarten sowie Immunzellen (T-Zelle, B-Zelle, Makrophage, Histiozyt)
- **Warum dies wichtig ist:** Repräsentiert die gemeinsamen molekularen Merkmale verschiedener Krebsarten
- **Klinische Relevanz:** Hilft pan-Krebs therapeutische Ziele zu identifizieren

#### **HBR (Human Brain Reference) = NORMALGEWEBE-KONTROLLEN**
- **Zusammensetzung:** Gesamt-RNA aus Gehirnen von 23 gesunden Kaukasiern, meist 60-80 Jahre alt
- **Warum Hirngewebe:** Bietet eine normale Gewebebasis zum Vergleich
- **Klinische Relevanz:** Zeigt, wie "gesunde" Genexpression aussieht

### **Die biologische Hypothese:**
**Krebszellen werden systematische Veränderungen in der Genexpression im Vergleich zu normalem Gewebe zeigen und offenbaren:**
1. **Onkogene** (krebsfördernde Gene), die überexprimiert sind
2. **Tumorsuppressoren** (krebsverhindernde Gene), die zum Schweigen gebracht werden
3. **Stoffwechselwege**, die verändert wurden, um Krebswachstum zu unterstützen
4. **Medikamentenziele**, die Krebszellen selektiv abtöten könnten

## 1 Datenexploration

### Count-Matrix verstehen

**Was ist eine Count-Matrix?**
Eine Count-Matrix ist die grundlegende Datenstruktur für RNA-seq-Analysen:
- **Zeilen**: Repräsentieren Gene oder Transkripte
- **Spalten**: Repräsentieren biologische Proben
- **Werte**: Rohzählungen, wie oft RNA-Fragmente jedes Gens sequenziert wurden

**Warum Count-Daten verwenden?**
- **Quantitative Messung**: Höhere Zählungen = höhere Genexpression
- **Biologische Replikate**: Mehrere Proben ermöglichen statistische Tests
- **Vergleichbarkeit**: Normalisierungsmethoden ermöglichen Vergleiche zwischen Proben

**Datenquelle:**
Diese Count-Matrix wurde durch Salmon-Quantifizierung von RNA-seq-Rohdaten erstellt:
1. **Rohdaten**: FASTQ-Dateien mit Millionen von RNA-Sequenz-Reads
2. **Quantifizierung**: Salmon ordnete Reads Genen zu und zählte sie
3. **Aggregation**: Zählungen wurden zu einer Matrix für die Analyse kombiniert

## 2 Differentielle Expressionsanalyse

### Differentielle Genexpression verstehen

**Was ist differentielle Expressionsanalyse?**
Die differentielle Expressionsanalyse identifiziert Gene, die in verschiedenen Bedingungen unterschiedlich exprimiert werden (z.B. Krebs vs. normal). Dies ist entscheidend für:
- **Biomarker-Entdeckung**: Gene, die Krebs diagnostizieren oder die Prognose vorhersagen können
- **Therapeutische Ziele**: Gene, die durch Medikamente angesteuert werden könnten
- **Mechanistisches Verständnis**: Wie sich Krebs auf molekularer Ebene von normalem Gewebe unterscheidet

**Wichtige statistische Konzepte:**

**Log2 Fold Change (Log2FC):**
- Misst, wie stark sich die Genexpression zwischen den Bedingungen ändert
- **Positive Werte** (z.B. +2): Gen ist in Krebs höher exprimiert (2² = 4-fach erhöht)
- **Negative Werte** (z.B. -2): Gen ist in Krebs niedriger exprimiert (2² = 4-fach verringert)
- **Null**: Keine Veränderung zwischen den Bedingungen

**P-Wert & Adjusted P-Wert:**
- **P-Wert**: Wahrscheinlichkeit, dass der beobachtete Unterschied zufällig ist
- **Adjusted P-Wert (padj)**: Korrigiert für Mehrfachtestprobleme beim Testen tausender Gene
- **Schwellenwert**: padj < 0,05 wird typischerweise als statistisch signifikant betrachtet

**Medizinische Interpretation:**
Ein Gen mit Log2FC = +3 und padj = 0,001 in Krebs bedeutet:
- Das Gen ist 2³ = 8-fach höher in Krebs exprimiert
- Diese Veränderung ist statistisch sehr signifikant
- Dies könnte ein Onkogen (krebsförderndes Gen) oder therapeutisches Ziel sein

**DESeq2-Methode:**
Wir verwenden DESeq2, ein statistisches Standardpaket für RNA-seq-Analysen:
1. **Normalisierung**: Berücksichtigt Unterschiede in der Bibliotheksgröße und Zusammensetzung
2. **Streuungsschätzung**: Modelliert biologische Variabilität zwischen Replikaten
3. **Statistischer Test**: Negative Binomialtests für differentielle Expression
4. **Mehrfachtestkorrektur**: Benjamini-Hochberg-Methode zur Kontrolle falscher Entdeckungen


### DESeq2-Analyse ausführen

Dieser Schritt führt die Hauptanalyse durch:
1. **Normalisierung**: Berücksichtigt Unterschiede in der Bibliotheksgröße zwischen Proben
2. **Streuungsschätzung**: Modelliert, wie konsistent die Expression über Replikate hinweg ist
3. **Statistischer Test**: Identifiziert Gene mit signifikanten Ausdrucksunterschieden
4. **Multiple Testkorrektur**: Berücksichtigt, dass wir Tausende von Genen gleichzeitig testen

**Erwartete Ausgabe:**
- Liste aller getesteten Gene
- Log2 Fold Change für jedes Gen (Krebs vs. Normal)
- Statistische Signifikanzwerte (p-Wert und adjusted p-Wert)
- Basismean-Expression (durchschnittliches Niveau über alle Proben)

**Spaltenerklärungen:**

### Signifikante Gene filtern

**Was sind "signifikante" Gene?**
Wir wenden zwei Kriterien an, um klinisch relevante Gene zu identifizieren:

1. **Statistische Signifikanz**: padj < 0,05
   - Falsch-Positiv-Rate (FDR) unter 5%
   - Die Veränderung ist unwahrscheinlich aufgrund von Zufall

2. **Biologische Signifikanz**: |log2FoldChange| > 1
   - Mindestens 2-fache Veränderung (2¹ = 2)
   - Große Veränderungen sind wahrscheinlicher biologisch bedeutsam
   - Potentielle therapeutische Ziele benötigen substanzielle Veränderungen

**Krebsgen-Klassifikation:**
- **Onkogene**: log2FC > +1 & padj < 0,05 (höher in Krebs)
  - Krebsfördernde Gene, die überaktiv sind
  - Potentielle Ziele für Inhibitoren
  
- **Tumorsuppressoren**: log2FC < -1 & padj < 0,05 (niedriger in Krebs)
  - Krebsverhindernde Gene, die inaktiviert sind
  - Potentielle Ziele für Reaktivierungstherapien

**Medizinische Relevanz:**
Diese gefilterten Gene repräsentieren:
- Potentielle diagnostische Biomarker
- Kandidaten für therapeutisches Targeting
- Biologische Prozesse, die bei Krebs verändert sind


```python
%%bash
awk 'NR==1 || ($7 < 0.05 && ($3 > 1 || $3 < -1))' deseq2_results.tsv > degs.tsv
```

## 3 Pathway-Anreicherungsanalyse

### Pathway-Analyse verstehen

**Warum Pathway-Analyse?**
Einzelne Gene arbeiten nicht isoliert – sie funktionieren in koordinierten biologischen Wegen (Pathways). Die Pathway-Analyse hilft uns zu verstehen:
- **Biologische Prozesse**: Was diese Krebsgene tun
- **Mechanistische Einsichten**: Wie Krebs auf Systemebene funktioniert
- **Therapeutische Strategien**: Welche Signalwege mit Medikamenten angesteuert werden könnten

**Was ist ein Pathway?**
Ein Pathway ist eine Gruppe von Genen, die zusammenarbeiten, um eine biologische Funktion auszuführen:
- **Beispiel**: "Zellzyklus-Pathway" = Gene, die die Zellteilung kontrollieren
- **Krebsrelevanz**: Krebszellen teilen sich oft unkontrolliert aufgrund von Zellzyklus-Pathway-Störungen

**Anreicherungsanalyse erklärt:**
Wenn viele Gene aus einem bestimmten Pathway in unserer Liste differentiell exprimierter Gene erscheinen, ist dieser Pathway wahrscheinlich wichtig für Krebs.

**Statistische Methode:**
- **Input**: Liste von 100 differentiell exprimierten Genen
- **Frage**: Sind mehr Gene aus einem bestimmten Pathway vorhanden, als zufällig erwartet?
- **Test**: Hypergeometrischer Test oder Fisher's Exact Test
- **Output**: P-Werte, die anzeigen, welche Pathways signifikant angereichert sind

**Pathway-Datenbanken:**

**Gene Ontology (GO):**
- **GO Biological Process (GO:BP)**: Was Gengruppen tun (z.B. "Zellproliferation")
- **GO Molecular Function (GO:MF)**: Welche Aktivitäten Gene haben (z.B. "Kinaseaktivität")
- **GO Cellular Component (GO:CC)**: Wo Gene funktionieren (z.B. "Nukleus")

**KEGG (Kyoto Encyclopedia of Genes and Genomes):**
- Kuratierte Datenbank menschlicher Signalwege
- Fokus auf Stoffwechsel- und Signalwege
- Nützlich für Medikamentenziel-Identifikation

**Reactome:**
- Detaillierte kuratierte Pathway-Datenbank
- Spezialisiert auf menschliche Biologie
- Hierarchisch organisiert (vom allgemeinen zum spezifischen Pathway)

**Erwartete Krebs-Pathways:**

**Hochregulierte Pathways (Onkogene):**
1. **Zellzyklus-Progression**: Unkontrollierte Zellteilung
2. **DNA-Replikation**: Unterstützt schnelle Proliferation
3. **Metabolische Umprogrammierung**: Veränderte Energieproduktion
4. **Angiogenese**: Bildung neuer Blutgefäße zur Tumorversorgung

**Herunterregulierte Pathways (Tumorsuppressoren):**
1. **Apoptose**: Programmierter Zelltod ist deaktiviert
2. **DNA-Reparatur**: Ermöglicht Mutationsansammlung
3. **Zellzyklus-Checkpoints**: Normale Wachstumshemmer sind inaktiviert
4. **Immunüberwachung**: Umgehung der Immunerkennung

**Klinische Anwendung:**
Pathway-Ergebnisse informieren über:
- **Medikamentenauswahl**: Targeting spezifischer Pathways (z.B. CDK-Inhibitoren für Zellzyklus)
- **Kombinationstherapie**: Targeting mehrerer Pathways gleichzeitig
- **Resistenzmechanismen**: Alternative Pathways, die Krebszellen nutzen könnten
- **Biomarker-Validierung**: Bestätigung, dass Gene biologisch zusammenhängend sind


### Pathway-Anreicherung durchführen

**Was macht dieser Schritt:**
1. **Nimmt unsere differentiell exprimierten Gene**: Liste signifikanter Gene aus DESeq2
2. **Fragt mehrere Pathway-Datenbanken ab**: GO, KEGG, Reactome
3. **Identifiziert angereicherte Pathways**: Welche biologischen Prozesse betroffen sind
4. **Liefert statistische Evidenz**: P-Werte für jede Pathway-Anreicherung

**Erwartete Laufzeit**: ~1-2 Minuten (abhängig von der Internetgeschwindigkeit)

**Ausgabedatei**: pathway_results.tsv mit:
- Pathway-Name und ID
- P-Wert und adjusted p-Wert
- Anzahl der Gene im Pathway
- Welche unserer Gene im Pathway sind

## 4 Drug Repositioning

### Drug Repositioning in der Krebsforschung

**Was ist Drug Repositioning?**
Drug Repositioning (auch Drug Repurposing genannt) beinhaltet die Entdeckung neuer therapeutischer Anwendungen für bestehende Medikamente. Dieser Ansatz bietet mehrere Vorteile:

1. **Schnellere Entwicklung**: 5-10 Jahre vs. 15-20 Jahre für neue Medikamente
2. **Bekannte Sicherheitsprofile**: Bestehende Medikamente haben etablierte Sicherheitsdaten
3. **Niedrigere Kosten**: Reduziert das Risiko und die Kosten der Medikamentenentwicklung
4. **Sofortige klinische Anwendung**: Kann in einigen Fällen off-label verschrieben werden

**Wie funktioniert computergestütztes Drug Repositioning?**
Unser Ansatz verwendet Genexpressionssignaturen:
1. **Krebssignatur**: Listen von Onkogenen (hochreguliert) und Tumorsuppressoren (herunterreguliert)
2. **Medikamentenwirkungsdatenbank**: Wie Tausende von Medikamenten die Genexpression beeinflussen
3. **Signatur-Matching**: Medikamente finden, die Krebsgenexpressionsmuster umkehren

**Erfolgsgeschichten:**
- **Metformin**: Diabetes-Medikament → Krebsprävention (200+ klinische Studien)
- **Aspirin**: Schmerzlinderung → Krebsprävention (FDA-zugelassen)
- **Rapamycin**: Immunsuppressivum → Krebs- und Altersforschung

**Unsere Skriptvorbereitung:**
Konvertiert unsere DESeq2-Ergebnisse in ein Format, das mit L1000CDS2 kompatibel ist, einer wichtigen Drug Repositioning-Datenbank.

### Abfrage der L1000CDS2-Medikamentendatenbank

**Was ist L1000CDS2?**
L1000CDS2 (L1000 Characteristic Direction Signature) ist ein computergestütztes Tool, entwickelt vom Ma'ayan-Labor, das:

1. **Datenbank-Umfang**: Enthält Genexpressionssignaturen für >20.000 Medikamente, getestet an humanen Zelllinien
2. **Signatur-Matching**: Verwendet mathematische Algorithmen, um Medikamente zu finden, die Krankheitssignaturen umkehren
3. **LINCS-Programm**: Teil der NIH Library of Integrated Network-based Cellular Signatures-Initiative

**Wie funktioniert der Algorithmus?**
1. **Eingabesignatur**: Unsere Krebsgensignatur (Onkogene + Tumorsuppressoren)
2. **Datenbanksuche**: Vergleicht mit medikamenteninduzierten Expressionsänderungen
3. **Bewertungssystem**: Berechnet, wie gut jedes Medikament die Krebssignatur umkehrt
4. **Ranking**: Gibt Medikamente zurück, gerankt nach ihrem Potenzial, Krebs entgegenzuwirken

**Interpretation der Ergebnisse:**
- **Negative Scores**: Medikamente, die Krebssignaturen umkehren (hohes therapeutisches Potenzial)
- **Positive Scores**: Medikamente, die Krebssignaturen nachahmen (diese vermeiden)
- **Score-Magnitude**: Größere absolute Werte deuten auf stärkere Effekte hin

**Klinische Validierung:**
Das System hat erfolgreich identifiziert:
- Bekannte Krebsmedikamente (validiert den Ansatz)
- Repositionierte Medikamente, die bereits in Krebsstudien sind
- Neuartige Repositionierungsmöglichkeiten für weitere Untersuchungen

### Warum Drug Repositioning funktioniert

**Wissenschaftliche Vorteile:**
- **Schnellere Entwicklungszeit**: 5-10 Jahre vs. 15-20 Jahre für neue Medikamente
